## Loading data from Google Drive

For example, if we want to read the local file `mydata.csv` to a notebook. We could do as following.

First, we need to upload the `csv` file to Google Drive.

Then, we need to call `drive` module from `google.colab` to mount our  Google Drive to the Colab environment.

Finally, we could just give the according path to the function like `pd.read_csv()` and load the data as a `pandas.DataFrame` object.


```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import pandas as pd
file_path = '/content/drive/My Drive/Colab Notebooks/mydata.csv'
df = pd.read_csv(file_path)
df.head()
```
